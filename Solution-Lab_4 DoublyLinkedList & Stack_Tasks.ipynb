**Assignment Part 1: Doubly Linked List**

For Assignment Part 1, you should write your full code in the following cells along with your driver codes on your own.

```python
#Assignment Part 1
class Patient:
  def __init__(self,id,name,age,bloodgroup):
    self.id=id
    self.name=name
    self.age=age
    self.bloodgroup=bloodgroup
class DoublyNode:
  def __init__(self,elem,next=None,prev=None):
    self.elem=elem
    self.next=next
    self.prev=prev

```

```python
class WRM:
    def __init__(self):
        self.head = DoublyNode(None)  # Dummy node
        self.head.next = self.head
        self.head.prev = self.head
    def registerPatient(self, id, name, age, bloodgroup):
        new_patient = Patient(id, name, age, bloodgroup)
        new_node = DoublyNode(new_patient)
        tail = self.head.prev
        tail.next = new_node
        new_node.prev = tail
        new_node.next = self.head
        self.head.prev = new_node
    def servePatient(self):
        if self.head.next == self.head:
            return "No patients in the waiting room"
        served_node = self.head.next
        served_patient = served_node.elem
        self.head.next = served_node.next
        served_node.next.prev = self.head
        return f"Serving Patient : {served_patient.name}"
    def cancelAll(self):
        self.head.next = self.head
        self.head.prev = self.head
        return "All appointments canceled."
    def canDoctorGoHome(self):
        return self.head.next == self.head
    def showAllPatients(self):
        if self.head.next == self.head:
            return ("No patients in the waiting room")
        current = self.head.next
        while current != self.head:
            patient = current.elem
            return f"Name: {patient.name} : ID: {patient.id}"
            current = current.next
    def reverseTheLine(self):
        if self.head.next == self.head:
            return "No patients in the waiting room"
        current = self.head.next
        while current != self.head:
            current.next, current.prev = current.prev, current.next
            current = current.prev
        self.head.next, self.head.prev = self.head.prev, self.head.next
        return "Line reversed successfully."
```

```python
#Write a Tester Code in this cell
print("**Welcome to Waiting Room Management System**")
wrm = WRM()

while True:
    print("\nOptions:")
    print("1. Add Patient")
    print("2. Serve Patient")
    print("3. Show All Patients")
    print("4. Can Doctor Go Home?")
    print("5. Cancel All Appointments")
    print("6. Reverse The Line")
    print("7. Quit")

    n = input("Choose option :")
    if n == '1':
        id = input("Enter patient ID: ")
        name = input("Enter patient name: ")
        age = input("Enter patient age: ")
        bloodgroup = input("Enter patient blood group: ")
        wrm.registerPatient(id, name, age, bloodgroup)
        print("Patient added successfully!")

    elif n == '2':
         print(wrm.servePatient())


    elif n == '3':
      print("Patients in sequence to serve: ")
      print(wrm.showAllPatients())


    elif n == '4':
        if wrm.canDoctorGoHome():
            print("Yes, the doctor can go home.")
        else:
            print("No, there are patients waiting.")

    elif n == '5':
        wrm.cancelAll()
        print("All appointments canceled.")

    elif n == '6':
        print(wrm.reverseTheLine())


    elif n == '7':
        print("Thank you")
        break

    else:
        print("Invalid choice. Please select a valid option.")
```

**Assignment Part 2: Stack**

Linked List based Stack is implemented in the following cell.

```python
class Node:
  def __init__(self,elem=None,next=None):
    self.elem = elem
    self.next = next

class Stack:
  def __init__(self):
    self.__top = None

  def push(self,elem):
    nn = Node(elem,self.__top)
    self.__top = nn

  def pop(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    e = self.__top
    self.__top = self.__top.next
    e.next = None
    return e.elem

  def peek(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    return self.__top.elem

  def isEmpty(self):
    return self.__top == None
```

```python
def diamond_count(stack,string):
  #TO DO
  count = 0
  for i in string:
    if i=="<":
      stack.push(">")
    else:
      elem = stack.peek()
      if stack.isEmpty():
        pass

      if i==elem:
        count+=1
        stack.pop()
  return count




print('Test 01')
stack= Stack()
string = '<..><.<..>> '
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 3
unittest.output_test(returned_value, 3)
print('-----------------------------------------')


print('Test 02')
stack = Stack()
string = '<<<..<......<<<<....>'
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 1
unittest.output_test(returned_value, 1)
print('-----------------------------------------')


print('Test 03')
stack = Stack()
string = '>>><...<<..>>...>...>>>'
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 3
unittest.output_test(returned_value, 3)
print('-----------------------------------------')
```

Task 2: Tower of Blocks

```python
def remove_block(stack, n):
  array = Stack()
  count = 1
  while stack.peek()!=None:
    if count==n:
      stack.pop()
      break
    else:
      elem = stack.peek()
      array.push(elem)
      stack.pop()
    count+=1
  while array.peek()!=None:
    elem = array.peek()
    stack.push(elem)
    array.pop()
  return stack


print('Test 01')
st = Stack()
st.push(4)
st.push(19)
st.push(23)
st.push(17)
st.push(5)
print('Stack:')
print_stack(st)
print('------')
remove_block(st,2)
print('After Removal')
print_stack(st)
print('------')

print()
print('======================================')
print()

print('Test 02')
st = Stack()
st.push(73)
st.push(85)
st.push(15)
st.push(41)
print('Stack:')
print_stack(st)
print('------')
remove_block(st,3)
print('After Removal')
print_stack(st)
print('------')

print()
print('======================================')
print()
```
