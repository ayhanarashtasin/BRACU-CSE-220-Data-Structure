Task 1: Zigzag Walk

```python
def walk_zigzag(floor):
  row,col = floor.shape
  top=0
  left = 0
  bottom = row-1
  while left<col:
    for i in range(top,bottom+1,2):
      print(floor[i][left],end=" ")
    print()
    left+=1
    if left>col-1:
      break
    if row%2==0:
      for i in range(bottom,top,-2):
        print(floor[i][left],end=" ")
      print()
      left+=1
    else:
      for i in range(bottom-1,top,-2):
        print(floor[i][left],end=" ")
      print()
      left+=1


floor = np.array([[ '3' , '8' , '4' , '6' , '1'],
                  ['7' , '2' , '1' , '9' , '3'],
                  ['9' , '0' , '7' , '5' , '8'],
                  ['2' , '1' , '3' , '4' , '0'],
                  ['1' , '4' , '2' , '8' , '6']]
                )



print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3 9 1
# 1 2
# 4 7 2
# 4 9
# 1 8 6
print("-------------------------------------")
floor = np.array([[ '3' , '8' , '4' , '6' , '1'],
                  ['7' , '2' , '1' , '9' , '3'],
                  ['9' , '0' , '7' , '5' , '8'],
                  ['2' , '1' , '3' , '4' , '0']]
                )

print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3 9
# 1 2
# 4 7
# 4 9
# 1 8
```

Task 2: Row Rotation Policy of BRACU Classroom

```python
def reverse_Matrix(matrix):
  i = 0
  j = len(matrix)-1
  while i<j:
    value = 0
    while value<len(matrix[0]):
      matrix[i][value],matrix[j][value] = matrix[j][value],matrix[i][value]
      value+=1
    i+=1
    j-=1
  for i in range(len(matrix)):
    j = 0
    k = len(matrix[0])-1
    while j<k:
      matrix[i][j],matrix[i][k] = matrix[i][k],matrix[i][j]
      j+=1
      k-=1
  return matrix
matrix = np.array([
[14,  8,  0,  4],
[9,  8,  13,  13],
[9,  3,  1,  4],
[2,  10,  13,  6]
])
print_matrix(matrix)
print('################')
print()
reversed_matrix = reverse_Matrix(matrix)
print_matrix(reversed_matrix)

#This should print
#|  6  |  13 |  10 |  2  |
#-------------------------
#|  4  |  1  |  3  |  9  |
#-------------------------
#|  13  |  13  |  8 |  9 |
#-------------------------
#|  4 |  0  |  8  |  14  |
#-------------------------
```

Task 5: Matrix Compression

```python
def compress_matrix(mat):
    #TO DO
    row,col = mat.shape
    r = row//2
    c = col//2
    arr = np.array([[0]*c]*r)
    sum = 0
    rw = 0
    cl = 0
    for i in range(0,row,2):
      for j in range(0,col,2):
        sum = mat[i][j]+mat[i+1][j]+mat[i][j+1]+mat[i+1][j+1]
        arr[rw][cl] = sum
        cl+=1
      rw+=1
      cl=0
      sum=0
    return arr




matrix=np.array([[1,2,3,4],
                 [5,6,7,8],
                 [1,3,5,2],
                 [-2,0,6,-3]
                 ])
print_matrix(matrix)
print('################')
print()
returned_array=compress_matrix(matrix)
print_matrix(returned_array)
#This should print
#|  14  |  22 |
#--------------
#|  2  |  10  |
#--------------
```

Task 6: Game Arena

```python
def play_game(arena):
  #TO DO
  row,col = arena.shape
  sum=0
  for i in range(row):
    for j in range(col):
      if arena[i][j]!=0 and arena[i][j]%50==0:
        if i>0:
          if arena[i-1][j]==2:
            sum+=2
        if i<row-1:
          if arena[i+1][j]==2:
            sum+=2
        if j>0:
          if arena[i][j-1]==2:
            sum+=2
        if j<col-1:
          if arena[i][j+1]==2:
            sum+=2
  if sum>=10:
    print(f"Points Gained: {sum}. Your team has survived the game.")
  else:
    print(f"Points Gained: {sum}. Your team is out")
arena=np.array([[0,2,2,0],
                [50,1,2,0],
                [2,2,2,0],
                [1,100,2,0]
                ])
print_matrix(arena)
play_game(arena)
#This should print
#Points Gained: 6. Your team is out.

print(".....................")
arena=np.array([[0,2,2,0,2],
                [1,50,2,1,100],
                [2,2,2,0,2],
                [0,200,2,0,0]
                ])
print_matrix(arena)
play_game(arena)
#This should print
#Points Gained: 14. Your team has survived the game.
```
